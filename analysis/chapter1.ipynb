```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def get_random_choice():
     return np.random.choice(range(1, 7), 10000, replace=True, p=None) 

choice = get_random_choice()
print(choice)
np.mean(choice)
```

```python
plt.hist(choice)
```

```python
expect = [np.mean(get_random_choice()) for i in range(1, 100000)]
plt.hist(expect)
```

```python
pd.DataFrame({'기대값': expect})
```
